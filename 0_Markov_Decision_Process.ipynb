```python
import numpy as np
```

```python
# Define the state space
states = ["S1", "S2", "S3", "S4"]
```

```python
#Define the action space
actions = ["A", "B"]
```

```python
# Create the state transition probabilities (Here we use 1)
# given certain s and a , the probablity of going into s'
transition_probabilities = {
    'S1': {"A":"S2", "B":"S3"},
    'S2': {"A":"S1", "B":"S4"},
    'S3': {"A":"S4", "B":"S1"},
    'S4': {"A":"S3", "B":"S2"}
}
```

```python
# create the reward function.Here it is only based on state and action
rewards = {
    ("S1", "A", "S2"): 10,
    ("S1", "B", "S3"): -10,
    ("S2", "A", "S1"): -1,
    ("S2", "B", "S4"): 5,
    ("S3", "A", "S4"): 7,
    ("S3", "B", "S1"): -2,
    ("S4", "A", "S3"): 3,
    ("S4", "B", "S2"): -8
}
```

```python
# Define the reward function R(s,a,s')
# s:current state
# s':next state
# a : acton take
def transition_probabilities_got(state, action, next_state):
  if transition_probabilities[state][action] == next_state:
    return 1
  else:
    return 0
```

```python
def get_reward(state, action, next_state):
  return rewards.get((state, action, next_state), 0)
```

```python
class MDP:
  def __init__(self, states, actions, transition_probabilities,
               reward_function, discount_factor=0.9):
    self.states = states
    self.actions = actions
    self.transition_probabilities = transition_probabilities_got
    self.reward_function = get_reward
    self.discount_factor = discount_factor
```

```python
mdp = MDP(states, actions, transition_probabilities_got, get_reward)
```

```python
transition_probabilities_got("S1", "A", "S2")
```

```python
reward = mdp.reward_function("S1", "A", "S2")
print(reward)
```
