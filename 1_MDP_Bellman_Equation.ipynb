*The relationship between Bellman Expectation Equation and the policy evaluation*

Bellman Expectation Equation = the rule
Value = reward now + discounted future value


Policy evaluation = the process
Use that rule-- Bellman Expecation Equation repeatedly to compute the value

# estimate the state value function with bellman expecation equation

```python
import numpy as np
```

```python
states = [1,2,3,4,5]
actions = ["up", "down"]
```

```python
# Define the transition probability and Rewards (Here the probability is 1 )
transition_probabilities = {
    1: {"up": 1, "down": 2},
    2: {"up": 1, "down": 3},
    3: {"up": 5, "down": 4},
    4: {"up": 1, "down": 1},
    5: {"up": 1, "down": 2}
}

rewards_function = {
    1: {"up": -2, "down": 2},
    2: {"up": -1, "down": -3},
    3: {"up": 0, "down": -4},
    4: {"up": 10, "down": 1},
    5: {"up": 5, "down": 0}
}
```

```python
# Define a policy , move up is prob 0.5, move down is prob 0.5
def policy(state):
  return {'up': 0.5, 'down': 0.5}
```

```python
# Define the policy evaluation function
def policy_evaluation(states, actions, transaction_probabilities, rewards_function, policy, gamma, num_iterations = 20):
  values = {state: 0 for state in states}

  for iteration in range(num_iterations):
    new_values = {}
    for state in states:
      new_value = 0
      for action in actions:
        # Calculate the expectation values of action based on the policy
        expected_value = sum(transition_probabilities[state][action] * (rewards_function[state][action] + gamma * values[next_state]) for next_state in states)
        new_value += policy(state)[action] * expected_value
      new_values[state] = new_value
    values = new_values # Update value_function for the next iteration

  return values
```

```python
# Estimate the state value functions with policy evaluation
estimated_values= policy_evaluation(states, actions, transition_probabilities, rewards_function, policy,gamma = 0.9, num_iterations=20)
```

```python
# Estimated State Values
for state, value in estimated_values.items():
  print(f"V({state}) = {value: .2f}")
```
