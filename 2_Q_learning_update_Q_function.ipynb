# What is Q-learning?

Q learning is a reinforcement learning method based on bellman equation. It is used in MDP to find the optiaml policy . In this process, Bellman equation is used for estimating the importance of future reward, and how to update the Q value base don the reward now , the estimated future reward as well

## Action Value Function of Q learning

Q learning is to learn the action value function Q(s,a), also called Q function or Q value function.

Q(s,a) The expected reward (accumulated reward) in the state to do action
* s: certain state
* a: action

Q learning find the optimal policy through update and optimize the Q value iteratively with Bellman Equation

**Q(s,a) <-- Q(s,a) + a * [R(s,a) + gamma * max(Q(s', a')) - Q(s,a)]**

* Q(s,a) --> Q value at state s, with action a
* a learning rate, how much it will update every time
* R(s,a) the immediate reward you can get after doing a in the state of s
* gamma --> discount factor, how important is the future reward
* max(Q(s', a')) --> choose the action a which can take the biggest q value in the next state


The aim is to choose the action which can take the highest Q in each state, in this way, it can get the biggest accumulated reward in the whole process

The most intelligent points for RL are (based on Q learning):
* Delay satisfication --> I care not only now, but also future
* I don't need to win every time point, but I want to win at last.
* Trial and Error No one teaches me, there is only grades for my action, I tried to find the best action (this is the way that i learn)


**Update the Q value function**

```python
import numpy as np
```

```python
num_states = 5 # Define the state space
num_actions = 2 # Define the action space
q_table = np.zeros((num_states, num_actions)) # Initialize the Q value as 0 at the beginning

learning_rate = 0.1 # Set the learning rate
discount_factor = 0.9 # Set the discount factor
num_epidoes = 10 # Set the number of episodes
```

```python
# Q leanring algorithm

for episode in range(num_epidoes):
    state = 0 # initial state
    done = False # the mark of game over

    while not done:
      # choose the action -- epsilon strategy
      epsilon = 0.1 # this part, I use it to explore
      if np.random.rand() < epsilon:
        action = np.random.randint(num_actions) # pick up the actions randomly
      else:
        action = np.argmax(q_table[state, :]) # pick up the action with the highest Q value

      # execute the acton and observe the reward
      if state == num_states - 1: # get the final state
          reward = 1
          done = True
      else:
        reward = 0

      # choose the next state:
      next_state = state + 1 if not done else state

      # Update the Q value with Bellman Equation
      q_table[state, action] = (1 - learning_rate) * q_table[state, action] + learning_rate * (reward + discount_factor * np.max(q_table[next_state, :]))
      state = next_state # update the state

print(q_table)
```
